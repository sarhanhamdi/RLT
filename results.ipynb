```python
from Models.rlt_regressor_advanced import RLTRegressorRAdvanced
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np

print("=" * 70)
print("TESTING ADVANCED RLTRegressorR")
print("=" * 70)
print()

# Create test data
X, y = make_regression(
    n_samples=200,
    n_features=20,
    n_informative=10,
    random_state=42
)

print(f"Dataset: X.shape={X.shape}, y.shape={y.shape}")
print(f"y range: [{y.min():.2f}, {y.max():.2f}]")
print()

# ============================================================================
# Test 1: Basic Model (Baseline)
# ============================================================================
print("=" * 70)
print("TEST 1: BASIC MODEL (Baseline)")
print("=" * 70)

model_basic = RLTRegressorRAdvanced(
    n_estimators=50,
    combsplit=1,
    muting=-1,
    muting_percent=0.0,
    reinforcement=False,
    random_state=42,
    verbose=True
)

model_basic.fit(X, y)
pred_basic = model_basic.predict(X)
mse_basic = mean_squared_error(y, pred_basic)
r2_basic = r2_score(y, pred_basic)
print(f"✅ MSE: {mse_basic:.4f}, R²: {r2_basic:.4f}")
print()

# ============================================================================
# Test 2: With Muting (30%)
# ============================================================================
print("=" * 70)
print("TEST 2: WITH MUTING (30%)")
print("=" * 70)

model_muting = RLTRegressorRAdvanced(
    n_estimators=50,
    muting=-1,
    muting_percent=0.3,
    random_state=42,
    verbose=True
)

model_muting.fit(X, y)
pred_muting = model_muting.predict(X)
mse_muting = mean_squared_error(y, pred_muting)
r2_muting = r2_score(y, pred_muting)
print(f"✅ MSE: {mse_muting:.4f}, R²: {r2_muting:.4f}")
print()

# ============================================================================
# Test 3: Linear Combinations (2 variables)
# ============================================================================
print("=" * 70)
print("TEST 3: LINEAR COMBINATIONS (2 variables)")
print("=" * 70)

model_combo2 = RLTRegressorRAdvanced(
    n_estimators=50,
    combsplit=2,
    random_state=42,
    verbose=True
)

model_combo2.fit(X, y)
pred_combo2 = model_combo2.predict(X)
mse_combo2 = mean_squared_error(y, pred_combo2)
r2_combo2 = r2_score(y, pred_combo2)
print(f"✅ MSE: {mse_combo2:.4f}, R²: {r2_combo2:.4f}")
print()

# ============================================================================
# Test 4: Linear Combinations (3 variables)
# ============================================================================
print("=" * 70)
print("TEST 4: LINEAR COMBINATIONS (3 variables)")
print("=" * 70)

model_combo3 = RLTRegressorRAdvanced(
    n_estimators=50,
    combsplit=3,
    random_state=42,
    verbose=True
)

model_combo3.fit(X, y)
pred_combo3 = model_combo3.predict(X)
mse_combo3 = mean_squared_error(y, pred_combo3)
r2_combo3 = r2_score(y, pred_combo3)
print(f"✅ MSE: {mse_combo3:.4f}, R²: {r2_combo3:.4f}")
print()

# ============================================================================
# Test 5: Muting + Linear Combinations
# ============================================================================
print("=" * 70)
print("TEST 5: MUTING (50%) + LINEAR COMBINATIONS (3)")
print("=" * 70)

model_advanced = RLTRegressorRAdvanced(
    n_estimators=50,
    combsplit=3,
    muting=-1,
    muting_percent=0.5,
    protect=5,
    random_state=42,
    verbose=True
)

model_advanced.fit(X, y)
pred_advanced = model_advanced.predict(X)
mse_advanced = mean_squared_error(y, pred_advanced)
r2_advanced = r2_score(y, pred_advanced)
print(f"✅ MSE: {mse_advanced:.4f}, R²: {r2_advanced:.4f}")
print()

# ============================================================================
# Test 6: Reinforcement Learning
# ============================================================================
print("=" * 70)
print("TEST 6: REINFORCEMENT LEARNING")
print("=" * 70)

model_reinforce = RLTRegressorRAdvanced(
    n_estimators=50,
    reinforcement=True,
    combsplit=2,
    muting=-1,
    muting_percent=0.3,
    random_state=42,
    verbose=True
)

model_reinforce.fit(X, y)
pred_reinforce = model_reinforce.predict(X)
mse_reinforce = mean_squared_error(y, pred_reinforce)
r2_reinforce = r2_score(y, pred_reinforce)
print(f"✅ MSE: {mse_reinforce:.4f}, R²: {r2_reinforce:.4f}")
print()

# ============================================================================
# Comparison Table
# ============================================================================
print("=" * 70)
print("COMPARISON TABLE")
print("=" * 70)

results = pd.DataFrame({
    'Model': [
        'Basic',
        'Muting 30%',
        'Combo (2-var)',
        'Combo (3-var)',
        'Combo (3) + Muting (50%)',
        'Reinforcement'
    ],
    'MSE': [
        f"{mse_basic:.4f}",
        f"{mse_muting:.4f}",
        f"{mse_combo2:.4f}",
        f"{mse_combo3:.4f}",
        f"{mse_advanced:.4f}",
        f"{mse_reinforce:.4f}"
    ],
    'R²': [
        f"{r2_basic:.4f}",
        f"{r2_muting:.4f}",
        f"{r2_combo2:.4f}",
        f"{r2_combo3:.4f}",
        f"{r2_advanced:.4f}",
        f"{r2_reinforce:.4f}"
    ]
})

print(results.to_string(index=False))
print()

# ============================================================================
# Feature Importance
# ============================================================================
print("=" * 70)
print("FEATURE IMPORTANCE")
print("=" * 70)

vi = model_advanced.get_feature_importance()
if vi is not None:
    print(f"Variable Importance Shape: {vi.shape}")
    print(f"Top 5 Features: {np.argsort(vi)[-5:]}")
    print(f"Top 5 Importance: {vi[np.argsort(vi)[-5:]]}")
else:
    print("Feature importance not available")
print()

# ============================================================================
# Prediction Examples
# ============================================================================
print("=" * 70)
print("PREDICTION EXAMPLES (First 5 Samples)")
print("=" * 70)

comparison_df = pd.DataFrame({
    'True Value': y[:5],
    'Basic Pred': pred_basic[:5],
    'Muting Pred': pred_muting[:5],
    'Combo3 Pred': pred_combo3[:5],
    'Advanced Pred': pred_advanced[:5]
})

print(comparison_df.to_string(index=False))
print()

print("=" * 70)
print("✅ ALL ADVANCED REGRESSION TESTS COMPLETED!")
print("=" * 70)

from Models.rlt_regressor_advanced import RLTRegressorRAdvanced

model = RLTRegressorRAdvanced(
    n_estimators=100,
    muting=-1,
    muting_percent=0.3,
    random_state=42
)

model.fit(X_train, y_train)
predictions = model.predict(X_test)


### Example 2: Regression with Linear Combinations

model = RLTRegressorRAdvanced(
    n_estimators=100,
    combsplit=3,  # 3-variable combinations
    random_state=42
)

model.fit(X_train, y_train)
predictions = model.predict(X_test)


### Example 3: Advanced Regression

model = RLTRegressorRAdvanced(
    n_estimators=100,
    combsplit=3,
    muting=-1,
    muting_percent=0.5,
    reinforcement=True,
    protect=5,
    verbose=True,
    random_state=42
)

model.fit(X_train, y_train)
predictions = model.predict(X_test)
feature_importance = model.get_feature_importance()

from sklearn.metrics import mean_squared_error, r2_score

strategies = [
    {"name": "Basic", "combsplit": 1, "muting_percent": 0.0},
    {"name": "Muting 30%", "combsplit": 1, "muting_percent": 0.3},
    {"name": "Combo (2)", "combsplit": 2, "muting_percent": 0.0},
    {"name": "Combo (3) + Muting", "combsplit": 3, "muting_percent": 0.5},
]

results = []
for strategy in strategies:
    model = RLTRegressorRAdvanced(**strategy, n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    
    results.append({
        'Strategy': strategy['name'],
        'MSE': mean_squared_error(y_test, pred),
        'R²': r2_score(y_test, pred)
    })

results_df = pd.DataFrame(results)
print(results_df)
```
